- BOOLEAN FILTERING


```python
x=np.random.randint(1,100,50)
print(x)
```

```python
x<10  #^ Boolean filtering
```

```python
x[x<10]
```

```python
f1=x>=20
f1
```

```python
f2=x<=30
f2
```

```python
print(x[f1 & f2])
print(x[(x>=20)&(x<=30)])
```
